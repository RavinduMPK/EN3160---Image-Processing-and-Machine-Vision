```python
# Spatial Filtering (Averaging) Using filter 2D

%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread(r"E:\UOM\Academic\Semester 5\Image processing and machine vision\Landscape-Color.jpg", cv.IMREAD_REDUCED_GRAYSCALE_2)

kernel  = np.ones((3,3), np.float32)/9
imgc = cv.filter2D(img,-1,kernel)

fig, axes = plt.subplots(1,2, sharex='all', sharey='all', figsize = (18,18))
axes[0].imshow(img, cmap = 'gray')
axes[0].set_title('Original')
axes[0].set_xticks([]), axes[0].set_yticks([])
axes[1].imshow(imgc,cmap='gray')
axes[1].set_title('Averaging')
axes[1].set_xticks([]), axes[1].set_yticks([])
plt.show()
```
